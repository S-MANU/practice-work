lets remove the unwanted columns.

```python
df1 = df.drop(['area_type', 'availability', 'society','balcony'], axis=1)
df1.columns
```

```python
df1= df1.dropna()
df1.head()
```

Extract Number of BHKs from size Column

```python
df1['bhk'] = df1['size'].apply(lambda x: int(x.split(' ')[0]))
```

```python
df1['bhk'].unique()
```

Handling square feet column

```python
def convert_sqft_to_num(x):
  try:
    if '-' in x:
      tokens = x.split('-')
      return (float(tokens[0]) + float(tokens[1])) / 2
    elif x.endswitch('Sq. Meter') or x.endswith('Acres') or x.endswith('Perch') or x.endswith('Sq. Yards'):
      return None
    else:
      return float(x)
  except:
    return None

df1['total_sqft'] = df1['total_sqft'].apply(convert_sqft_to_num)
df1 = df1.dropna(subset=['total_sqft'])
```

```python
df1.head()
```

Handling the price

```python
df1['price_per_sqft'] = (df1['price']*100000) / df1['total_sqft']
```

```python
df1.head()
```

Cleaning location

```python
df1['location'] = df1['location'].apply(lambda x: x.strip())
location_stats = df1['location'].value_counts()
location_stats_less_than_10 = location_stats[location_stats <= 10]
df1['location'] = df1['location'].apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
```

```python
df1.describe()
df1.head()
```

```python
df1 = df1.drop(['size'], axis=1)
```

```python
df1.head()
```

Day 3

2 BHK costing same price of 3 bhk in same location that is suspecious so filter this kind outliers.

```python
import numpy as np
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df['price_per_sqft']),
                'std': np.std(bhk_df['price_per_sqft']),
                'count': bhk_df.shape[0]
            }

        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(
                    exclude_indices,
                    bhk_df[bhk_df['price_per_sqft'] < stats['mean']].index.values
                )

    return df.drop(exclude_indices, axis=0)

df2 = remove_bhk_outliers(df1)


```

Removing bathroom outliers, anything more than 4 bathroom is not realistic.

```python
df3 = df2[df2['bath']  < df2['bhk'] + 2]
```

clean data to use and train

```python
df4 = df3.copy()
df4 = df4.drop(['price_per_sqft'], axis=1)
```

ML model don't understand location like 'Whitefield', so

```python
dummies = pd.get_dummies(df4['location'])
df5 = pd.concat([df4.drop('location', axis=1),dummies.drop('other', axis = 1)], axis=1)
```

day 4: Model Training and Evolution

Split the data

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df5.drop('price', axis=1)
y = df5['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

Trining the model

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

Predict Function:

lets say we want to predict:
  2 BHK in whitefield , 1200sqft, 2 bath

```python
def predict_price(location, sqft, bath, bhk):
    loc_index = np.where(X.columns == location)[0][0] if location in X.columns else -1

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    x_df = pd.DataFrame([x], columns=X.columns)

    return model.predict(x_df)[0]
```

```python
predict_price('Whitefield', 1200, 2, 2)
```
